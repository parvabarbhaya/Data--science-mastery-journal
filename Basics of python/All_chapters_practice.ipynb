```python
# Take a list of numbers  as an user input and print their sum and average.
user_inpt = input("insert the list of numbers (separated by space or comma): ")
tokens = user_inpt.replace(',', ' ').split()
numbers = [float(x) for x in tokens] if tokens else []
num_sum = 0.0
for t in tokens:
    try:
        num_sum += float(t)
    except ValueError:
        # ignore non-numeric tokens
        pass
avg_of_numbers = num_sum / len(numbers) if numbers else 0
print(f"the sum is {num_sum} and the average is {avg_of_numbers}")
```
