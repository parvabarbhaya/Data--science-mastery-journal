Q9 — Condition Ordering | Write the Code
Write a function grade(score) that returns:

* "Invalid" if score is below 0 or above 100
* "A" if 90–100
* "B" if 75–89
* "C" if 50–74
* "F" if below 50

```python
def grade(s):
    if (s < 0) or (s >100):
        return "invalid"
    elif (s >=90) & (s <= 100):
        return "A"
    elif (s >=75) & (s <= 89):
        return "B"
    elif (s >=50) & (s <= 74):
        return "C"
    elif (s<50):
        return "F"
score_checker = [-5, 0, 49, 50, 75, 90, 100, 105]
for i in score_checker:
    print(grade(i))
```
